```python
# Now we are having IQR will check the shape of outliers
upperLimit=Q3+(1.5*IQR)
lowerLimit=Q1-(1.5*IQR)
print(upperLimit,lowerLimit)
df_outlier=df_cleaned[(df_cleaned["annual_inc"]<lowerLimit) | (df_cleaned["annual_inc"]>upperLimit)]
```

```python
df_outlier["annual_inc"].shape
```

##### *Hence there are 1117 outliers in annual_income colunm will handle outliers either will remove it or replacing it with the nearest values fro the annual income I am removing outliers because they are having high chance of repaying* 

```python
df_cleaned=df_cleaned[(df_cleaned["annual_inc"]<=lowerLimit) | (df_cleaned["annual_inc"]<=upperLimit)]
df_cleaned.shape
```
